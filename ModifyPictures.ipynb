```python
# recording_name_list stores the relative dir of each picture, such as ''./Picture/New/26classes/modificationPictures/door_wood_knock-1442.png'
# nameClasses stores the class name of each picture, sucha as 'dog'
recording_name_list = []
nameClasses = []
for item in range(len(fileNames)):
    if name_class[item].endswith('.png'):
        nameClasses.append(fileNames[item][:-9])        
        namedir = f'{parentDir}/{name_class[item]}'
        recording_name_list.append(namedir)
```

```python
# modify the pictures stored in parentDir and store them in the storeDir
# flip_1 will make the original picture updown 
# flip_2 will make the original picture left right, just like the image in the mirror. 
# we use '#' and '$' to distinguish each other and the original picture which is '-'
for item in range(661, len(recording_name_list)):
    image = mpimg.imread(recording_name_list[item])
    category = nameClasses[item]
    flip_1 =sess.run( tf.image.flip_up_down(image))
    mpimg.imsave(f'{storeDir}/{category}#{item:04d}.png',flip_1)
   
    flip_2 = sess.run(tf.image.flip_left_right(image))
    mpimg.imsave(f'{storeDir}/{category}${item:04d}.png',flip_2)
```
